```python
# imports
import os
import pandas as pd 
import random
```

```python
# variables
startdir = r'D:\git\FileCarver\Dataset\dbc'
outputdir = r'D:\git\FileCarver\Dataset\dbc_new'
file_names = []
keywords = ["VERSION", "NS_", "BO_", "BU_:","BS_:", "CM_", "VAL_"]
```

```python
# get all file paths

for path, subdir, files in os.walk(startdir):
    for file in files:
        file_names.append(startdir + "\\" + file)
file_names
```

```python
# load dbc file in dataframe

df = pd.DataFrame()

for file in file_names:
        with open(file, 'r') as f:
            dict_file = {i: [] for i in keywords}
            key_alt = ''
            str_new = ''
            for line in f:
                key =  line.split(' ', 1)[0]
                if key in keywords:
                    if not key_alt:
                        key_alt = key
                    if str_new:
                        dict_file[key_alt].append(str_new)
                        str_new = line
                    else:
                        str_new = str_new + line
                    key_alt = key
                else:
                    str_new = str_new + line
            df_dict = pd.DataFrame([dict_file])
            df = pd.concat([df, df_dict], ignore_index=True)
             
df
```

```python
# create new dbc files 
file_name = 'generated_dbc_file_'

for i in range(500):
    # create new file
    name = file_name + str(i)
    path = outputdir + '\\' + name + '.dbc'
    
    with open(path, 'w') as f:
    
        len_df = len(df)-1
    
        items = []
        for name in list(df.columns.values):
            randnum = random.randint(0, len_df)
            # for each column pick random row 
            listee = df[name][randnum]
            # pick a number of random objects in list 
            if listee:
                randnum = random.randint(len(listee)//2, len(listee)-1)
                items = random.choices(listee[1:], k=randnum)
                items.insert(0,listee[0])
            for line in listee:
                f.write(line)
        
```
